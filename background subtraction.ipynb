# Background subtraction evaluation

Opencv comes with a number of algorithms predefined


This notebook serves to do a first pass of evaluations of these different algorithms

```python
import cv2
```

```python
import glob
```

```python
videos = glob.glob("../../**/*mp4",recursive = True)
```

```python
class GenericVideo:
    vid = None
    cap = None
    frame = None
    def process(self):
        self.cap = cv2.VideoCapture(self.vid)
        while(1):
            ret,frame = self.cap.read()
            if not ret:
                break
            self.frame = frame
            self.doAdvanced()
            cv2.imshow("frame",self.frame)
            k = cv2.waitKey(30) &0xff
            if k == 27:
                break
        self.cap.release()
        cv2.destroyAllWindows()
    def doAdvanced(self):
        pass
```

```python
gv = GenericVideo()
gv.vid=  videos[0]
gv.process()
```

```python
class MOG1Test(GenericVideo):
    def doAdvanced(self):
        print("doing extra step")
        
test = MOG1Test()
test.vid = videos[0]
test.process()
```

## MOG1 background subtraction
The mog stands for mixture of gradients no doubt

This section of the code tests a simple background subtraction method and overrides the step where the generic video just passes 

preliminary evaluation
- seems quite fast
- visual artifacts are pretty diminished
- sometimes the sections aren't grouped exactly, as in it's pretty grainy
- it's good at detecting initial motions from completely static, but isn't outfitted well to continue to detect things that are still moving in the same area
    - for example when the soil continues to move down the same chute it isn't getting picked up
- start of video clip  
- middle of video clip 

```python
class MOG1(GenericVideo):
    fgbg = cv2.bgsegm.createBackgroundSubtractorMOG()
    def doAdvanced(self):
        fgmask = self.fgbg.apply(self.frame)
        cv2.imshow("mask",fgmask)
    
m1 = MOG1()
m1.vid = videos[0]
m1.process()
    
```

## GSOC background algo
This class employs a gsoc algorithm developed after 2014 to the background subtraction

prelim eval:
- prelim eval
- runs much slower
- has fatter outlines of background subtraction
- has bigger jumps
- could require messing with the apply parameters
- 

```python
class GSOC(MOG1):
    fgbg= cv2.bgsegm.createBackgroundSubtractorGSOC()
    
gsoc1 = GSOC()
gsoc1.vid  = videos[0]
gsoc1.process()
```

## MOG2

eval

```python
class MOG2(MOG1):
    fgbg= cv2.createBackgroundSubtractorMOG2()
    
m2 = MOG2()
m2.vid  = videos[0]
m2.process()
```
